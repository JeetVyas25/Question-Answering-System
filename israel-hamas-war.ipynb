````python
from torch import cuda, bfloat16
import transformers

model_id = 'meta-llama/Llama-2-7b-chat-hf'

device = f'cuda:{cuda.current_device()}' if cuda.is_available() else 'cpu'

bnb_config = transformers.BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type='nf4',
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=bfloat16
)

hf_auth = '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'
model_config = transformers.AutoConfig.from_pretrained(
    model_id,
    use_auth_token=hf_auth
)

model = transformers.AutoModelForCausalLM.from_pretrained(
    model_id,
    trust_remote_code=True,
    config=model_config,
    quantization_config=bnb_config,
    device_map='auto',
    use_auth_token=hf_auth
)

model.eval()

print(f"Model loaded on {device}")
````

````python
tokenizer = transformers.AutoTokenizer.from_pretrained(
    model_id,
    use_auth_token=hf_auth
)
````

````python
stop_list = ['\nHuman:', '\n```\n']

stop_token_ids = [tokenizer(x)['input_ids'] for x in stop_list]
stop_token_ids
````

````python
import torch

stop_token_ids = [torch.LongTensor(x).to(device) for x in stop_token_ids]
stop_token_ids
````

````python
from transformers import StoppingCriteria, StoppingCriteriaList

class StopOnTokens(StoppingCriteria):
    def __call__(self, input_ids: torch.LongTensor, scores: torch.FloatTensor, **kwargs) -> bool:
        for stop_ids in stop_token_ids:
            if torch.eq(input_ids[0][-len(stop_ids):], stop_ids).all():
                return True
        return False

stopping_criteria = StoppingCriteriaList([StopOnTokens()])
````

````python
generate_text = transformers.pipeline(
    model=model,
    tokenizer=tokenizer,
    return_full_text=True,
    task='text-generation',
    stopping_criteria=stopping_criteria, 
    temperature=0.1, 
    max_new_tokens=512,  
    repetition_penalty=1.1 
)
````

````python
res = generate_text("Explain me the difference between Data and information")
print(res[0]["generated_text"])
````

````python
from langchain.llms import HuggingFacePipeline

llm = HuggingFacePipeline(pipeline=generate_text)

llm(prompt="Explain me the difference between Data and information")
````

````python
import json
import re
from datetime import datetime

def clean_text(text):
    if text is None:
        return ""
    text = text.replace('\n', ' ')
    text = re.sub(r'[^A-Za-z0-9\s.,!?\'"]+', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def convert_to_qa_format(data):
    formatted_data = []
    for item in data:
        title = clean_text(item.get('title', ''))
        article_body = clean_text(item.get('articleBody', ''))
        date_modified = item.get('dateModified', {}).get('dateModified', '')
        
        if date_modified:
            try:
                date_modified = datetime.fromisoformat(date_modified.replace('Z', '+00:00')).isoformat()
            except ValueError:
                date_modified = ''
        
        formatted_item = {
            'prompt': f"Title: {title}\n\nArticle: {article_body}",
            'response': article_body
        }
        formatted_data.append(formatted_item)
    return formatted_data

qa_data = convert_to_qa_format(data)

with open('qa_data.jsonl', 'w') as outfile:
    for entry in qa_data:
        json.dump(entry, outfile)
        outfile.write('\n')
````

````python
import json

def read_jsonl_head(file_path, num_lines=5):
    data = []
    with open(file_path, 'r') as file:
        for _ in range(num_lines):
            line = file.readline()
            if not line:
                break
            json_obj = json.loads(line)
            data.append(json_obj)
    return data

file_path = '/kaggle/working/qa_data.jsonl'

jsonl_head = read_jsonl_head(file_path, num_lines=5)

for json_obj in jsonl_head:
    print(json.dumps(json_obj, indent=4))
    print(json_obj['response'])
````

````python
from langchain.document_loaders import WebBaseLoader

web_links = ['https://www.britannica.com/event/Israel-Hamas-War' , 'https://en.wikipedia.org/wiki/Israel%E2%80%93Hamas_war']
loader = WebBaseLoader(web_links)
documents1= loader.load()
````

````python
import json
from langchain.docstore.document import Document
from langchain.text_splitter import RecursiveCharacterTextSplitter

def read_jsonl(file_path):
    data = []
    with open(file_path, 'r') as file:
        for line in file:
            json_obj = json.loads(line)
            data.append(json_obj['response'])  # Assuming 'response' contains the text
    return data

file_path = '/kaggle/working/qa_data.jsonl'

json_data = read_jsonl(file_path)

documents = [Document(text) for text in json_data]

documents.extend(documents1)

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=20)

all_splits = text_splitter.split_documents(documents)


````

````python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS

model_name = "sentence-transformers/all-mpnet-base-v2"
model_kwargs = {"device": "cuda"}

embeddings = HuggingFaceEmbeddings(model_name=model_name, model_kwargs=model_kwargs)

vectorstore = FAISS.from_documents(all_splits, embeddings)
````

````python
from langchain.chains import ConversationalRetrievalChain

chain = ConversationalRetrievalChain.from_llm(llm, vectorstore.as_retriever(), return_source_documents=True)
````

````python
chat_history = []

query = "what happened at the Al-shifa hospital"
result = chain({"question": query, "chat_history": chat_history})

print(result['answer'])
````
